## 1) Qual é a média de idade do público masculino?
## 2) Qual é a média de idade do público feminino?

## 4) Quantas passageiras do sexo feminino sobreviveram?
## 5) Quem estava hospedado na cabine B42?

## 8) Qual foi a taxa que Jussila, Mr. Eiriik pagou?

## 10) Valor total pago pelos passageiros?



```python
# Exercício 1

generoM = titanic_df[titanic_df["Sex"] == "male"]

idade = generoM["Age"].mean()

print("A média de idade do público masculino é: {:.1f} anos".format(idade))
```

```python
# Exercício 2

generoF = titanic_df[titanic_df["Sex"] == "female"]

idade = generoM["Age"].mean()

print("A média de idade do público feminino é: {:.1f} anos".format(idade))
```

```python
# Exercício 3

qnt_pessoas = len(titanic_df)

print(f"Quantidade total de tripulantes no Titanic foi de: {qnt_pessoas} pessoas.")
```

```python
# Exercício 4

sobreviventesF= titanic_df[(titanic_df["Sex"] == "female") & (titanic_df["Survived"] == 1)]
total_mulheres = len(titanic_df[titanic_df["Sex"] == "female"])

qnt_vivas = len(sobreviventesF)

print(f"Quantidade total de mulheres no Titanic: {total_mulheres} sobreviventes.")
print(f"Quantidade de mulheres sobreviventes: {qnt_vivas} sobreviventes.")
```

```python
# Exercício 5

cabine_B42 = titanic_df[titanic_df["Cabin"] == "B42"]

hosp = cabine_B42["Name"].values[0]

print(f"Passageiro hospedado na cabine B42: {hosp}.")
```

```python
# Exercício 6

pessoa = titanic_df[titanic_df["Name"] == "Behr, Mr. Karl Howell"]

idade = pessoa["Age"].values[0]

print(f"A idade de Behr, Mr. Karl Howell era: {idade} anos")
```

```python
# Exercício 7

nome = titanic_df[titanic_df["Name"] == "Young, Miss. Marie Grice"]

cabine = nome["Cabin"].values[0]

print(f"Cabine de Young, Miss. Marie Grice: {cabine}.")
```

```python
# Exercício 8

nome_passageiro = titanic_df[titanic_df["Name"] == "Jussila, Mr. Eiriik"]

valor_pago = nome_passageiro["Fare"].values[0]

print("Valor pago pelo Jussila, Mr. Eiriik foi de: ${:,.3f}".format(valor_pago))
```

```python
# Exercício 9

nome_passageiro2 = titanic_df[titanic_df["Name"] == "Fynney, Mr. Joseph J"]

ticket = nome_passageiro2["Ticket"].values[0]

print(f"O ticket do Fynney, Mr. Joseph J era: {ticket}.")
```

```python
# Exercício 10
# ## 10) Valor total pago pelos passageiros?

valor_pago = titanic_df["Fare"].sum()

print("O valor total pego pelos passageiros do Titanic foi de: ${:,.2f} ".format(valor_pago))
```
